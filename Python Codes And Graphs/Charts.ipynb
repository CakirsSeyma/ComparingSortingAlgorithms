```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

data = {
    'Algorithm': ['Selection', 'Insertion', 'Shellsort', 'Merge', 'ThreeWay', 'Lomuto', 'Hoare', 'Heapsort'],
    'Random': [0.004, 0.006, 0.001, 0.002, 0.001, 0.001, 0.001, 0.001],
    'Sorted': [0.004, 0.00001, 0.00001, 0.002, 0.001, 0.013, 0.005, 0.001],
    'Reversed': [0.005, 0.006, 0.001, 0.001, 0.002, 0.01, 0.005, 0.001]
}

df = pd.DataFrame(data)

df_melted = df.melt('Algorithm', var_name='Type', value_name='Time')

plt.figure(figsize=(10, 6))
sns.barplot(x='Algorithm', y='Time', hue='Type', data=df_melted)
plt.title('Algorithm Performance 1K Size')
plt.ylabel('Time (s)')
plt.xlabel('Algorithm')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.tight_layout()

plt.show()
```
